```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k_true = 2
b_true = 4

num_points = 100
x = np.random.uniform(-10, 10, num_points)
y_true = k_true * x + b_true
noise = np.random.normal(0, 2, num_points)
y = y_true + noise
```

```python
def least_squares_fit(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    numerator = np.sum((x - x_mean) * (y - y_mean))  
    denominator = np.sum((x - x_mean) ** 2)         
    k = numerator / denominator
    
    b = y_mean - k * x_mean
    
    return k, b
```

```python
k_my, b_my = least_squares_fit(x, y)

k_polyfit, b_polyfit = np.polyfit(x, y, 1)

print("Базові параметри:")
print(f"  k = {k_true}, b = {b_true}")
print("\nМоя оцінка:")
print(f"  k = {k_my:}, b = {b_my:}")
print("\nОцінка polyfit:")
print(f"  k = {k_polyfit:}, b = {b_polyfit:}")
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(x, y, label="Згенеровані дані")
plt.plot(np.sort(x), k_true * np.sort(x) + b_true, color="green", label="Базова пряма")
plt.plot(np.sort(x), k_my * np.sort(x) + b_my, color="red", label="Моя оцінка")
plt.plot(np.sort(x), k_polyfit * np.sort(x) + b_polyfit, color="blue", label="Оцінка polyfit")
plt.legend()
plt.show()
```
